#### Age

```python
df.Age.describe()
```

I would consider Age is this analysis to be continous. There are a wide range of ages as opposed to age groups and age has a finite start at 0 and is precise.

```python
df[df.Age.isnull()]
```

Out of the 891 entries, 177 don't have an age. To replace this information, an average of all ages will be used.

Previous to filling in an average age for null ages.

```python
df.Age.describe()
```

```python
avgAge = df.Age.mean()
```

```python
df.Age = df.Age.fillna(value=avgAge)
```

After filling in an average age for null ages.

```python
df.Age.describe()
```

```python
fig, axs = plt.subplots(1,3)
df[df.Age < 10].Survived.value_counts().plot(kind='barh', title='Under 10 Survivorship', ax=axs[0])
df[(df.Age > 10) & (df.Age < 20)].Survived.value_counts().plot(kind='barh', title='10 - 20 Survivorship', ax=axs[1])
df[(df.Age > 20) & (df.Age < 30)].Survived.value_counts().plot(kind='barh', title='20 - 30 Survivorship', ax=axs[2])
```

```python
fig, axs = plt.subplots(1,3)
df[(df.Age > 30) & (df.Age < 40)].Survived.value_counts().plot(kind='barh', title='30 - 40 Survivorship', ax=axs[0])
df[(df.Age > 40) & (df.Age < 50)].Survived.value_counts().plot(kind='barh', title='40 - 50 Survivorship', ax=axs[1])
df[(df.Age > 50)].Survived.value_counts().plot(kind='barh', title='50 - 60 Survivorship', ax=axs[2])
```

Looking at the 6 graphs above, it seems like the only age group that had an increased survivorship was the under 10 age group. Which follows the theory that "women and children" were saved first.

```python
df.Age.value_counts().plot(kind='bar')
```

#### SibSp (Number of Siblings/Spouses)
